# Comparando dados de 1995 e 2022

```python
# agora, vamos comparar os dados de 1995 e 2022

# vamos começar com o arquivo de cursos, comparando as colunas presentes em cada um dos arquivos

# lista de colunas de cada arquivo
colunas_cursos_1995 = dados_1995_cursos.columns
colunas_cursos_2022 = dados_2022_cursos.columns

# agora, comparando os registros de cada lista com o nome das colunas
for i, coluna in enumerate(colunas_cursos_1995, start=1):
    if coluna not in colunas_cursos_2022:
        print(f'### ATENÇÃO! ### A coluna {coluna} não existe no arquivo de 2022 (linha {i})')
    else:
        print(f'A coluna {coluna} existe no arquivo de 2022 (linha {i})')

# agora, vamos adicionar uma linha para contar a quantidade de colunas que existem no arquivo de 2022 (somar os resultados do print acima)
print(f'A quantidade de colunas equivalentes entre os arquivos é de {len(colunas_cursos_2022)}')
```

```python
# agora, vamos comparar os dados das ie de 1995 e 2022:

# lista de colunas de cada arquivo
colunas_ie_1995 = dados_1995_ie.columns
colunas_ie_2022 = dados_2022_ie.columns

# agora, comparando os registros de cada lista com o nome das colunas
```

```python
# como apenas obtivemos o registro de colunas de 1995 que não estão presentes em 2022, vamos verificar o contrário:

for coluna in colunas_ie_2022:
    if coluna not in colunas_ie_1995:
        print(f'### ATENÇÃO! ### A coluna {coluna} não existe no arquivo de 1995')
    elif coluna in colunas_ie_1995:
        print(f'A coluna {coluna} existe no arquivo de 1995')
```
